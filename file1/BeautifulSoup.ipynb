# import required Libraries

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/more-than-30instant-discount')
```

```python
page
```

# page content

```python
soup = BeautifulSoup(page.content)
soup
```

# Scraping first Name

```python
#first we will use html_tag where we have the first title of the restaurants

first_title = soup.find('div',class_="restnt-info cursor")
first_title
```

```python
first_title.text
```

# Scraping first Location

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

# Scraping first price

```python
sta  = soup.find('span',class_="double-line-ellipsis")
sta.text
```

```python
sta  = soup.find('span',class_="double-line-ellipsis")
sta.text.replace('₹','')
```

# Scraping Multiple Titles

```python
# now we have all the tags in which there are the job titles
# now we will extract the text from these tags one by one looping these tags

titles = []

for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

# Scraping Multiple Locations

```python
Locations = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Locations.append(i.text)
    
Locations
```

# Scraping Multiple Price

```python
prices = []

for i in soup.find_all('span',class_="double-line-ellipsis"):
    prices.append(i.text)
    
prices
```

```python
prices = []

for i in soup.find_all('span',class_="double-line-ellipsis"):
    prices.append(i.text.replace('₹','$'))
    
prices
```

```python
images = []
for i in soup.find_all("img",class_="no-img"):
    images.append(i['data-src'])
    
images
```

```python
#printing length

print(len(titles),len(Locations),len(prices),len(images))
```

```python
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Location':Locations,'Price':prices,'Images':images})
df
```
